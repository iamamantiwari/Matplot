# Matplotlib Mutliple Bar Chart

```python
import matplotlib.pyplot as plt
```

# Simple Bar Chart showing revenues of major Us Tech Companies

```python
Company=['Google','Amazon','Microsoft','Facebook']
revenue=[90,136,89,27]
```

```python
plt.bar(Company,revenue)
plt.xlabel('Companay')
plt.ylabel('Revenue(Bln)')
plt.title('US Companies Stock')
for c,r in zip(Company,revenue):
    plt.text(c,r,r)
    
plt.show()    
```

```python
plt.barh(Company,revenue)
plt.xlabel('Companay')
plt.ylabel('Revenue(Bln)')
plt.title('US Companies Stock')

# for c,r in zip(Company,revenue):
#     plt.text(c,r,r)
    
plt.show()  
```

# Multiple Bars showing revenue and profit of major US tech Companies 

```python
company=['Google','Amazon','Microsoft','Facebook']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue,label='Revenue')
plt.bar(company,profit,label='Profit')
plt.xlabel('Company')
plt.ylabel('Revenue(bln)vs Profit')
plt.title('US Compnaies Stock')
plt.legend()
plt.show()
```

```python
import numpy as np
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.xticks(xpos,company)
# plt.xlabel('Company')
plt.ylabel('Revenue(bln)vs Profit')
plt.title('US Compnaies Stock')

for c,r in zip(xpos,revenue):
    plt.text(c-0.2,r+0.2,r)
    
for c,p in zip(xpos,profit):
    plt.text(c,p,p)

plt.legend()
plt.show()
```
