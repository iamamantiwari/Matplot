```python
years=[2018,2019,2020,2021,2022,2023]
sales=[1000000,1500000,600000,800000,1200000,1500000]
plt.plot(years,sales,c='b',linestyle='--')
plt.xticks(range(2018,2025,1))
plt.yticks(range(500000,2000000,200000))
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title('Sales report in Years',size=20,c='k')
plt.show()
```

```python
years=[2018,2019,2020,2021,2022,2023]
sales=[1000000,1500000,600000,800000,1200000,1500000]
plt.plot(years,sales,c='b',linestyle='solid',linewidth=2,marker='o',markerfacecolor='r',
         markeredgecolor='k',markersize=10,alpha=0.6) 
   #alpha= 0.0 to 1.0


plt.xticks(range(2018,2025,1),c='b',size=10)
plt.yticks(range(500000,2000000,200000))
plt.xlabel('Years',size=20)
plt.ylabel('Sales',size=20)
plt.title('Sales Report in Years',size=20,c='r')
plt.show()
```

```python
years=[2018,2019,2020,2021,2022,2023]
sales=[1000000,1500000,600000,800000,1200000,1500000]
plt.plot(years,sales,c='b',linestyle='solid',linewidth=2,marker='o',markerfacecolor='r',
         markeredgecolor='k',markersize=10,alpha=0.6) 
   #alpha= 0.0 to 1.0


plt.xticks(range(2018,2025,1),c='b',size=10)
plt.yticks(range(500000,2000000,200000))
plt.xlabel('Years',size=20)
plt.ylabel('Sales',size=20)
plt.title('Sales Report in Years',size=20,c='r')
plt.text(2021,1500000,'This is Line Chart',c='g',size=12)
for y,s in zip(years,sales):
    plt.text(y,s,s)
plt.show()
```
